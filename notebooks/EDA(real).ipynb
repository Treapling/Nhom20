```python
# Ô 1 — Import thư viện

import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# Ô 2 — Đọc dữ liệu CSV

df = pd.read_csv("/content/drive/MyDrive/data_public.csv")

df.head()
```

```python
# Ô 3 — Xem kích thước dataset

print("Shape:", df.shape)
```

```python
# Ô 4 — Xem thông tin tổng quát

df.info()
```

```python
# Ô 5 — Kiểm tra dữ liệu thiếu

df.isnull().sum()
```

```python
# Ô 6 — Xem thống kê mô tả

df.describe()
```

```python
# Ô 7 — Xem tên các cột

print(df.columns)
```

```python
# Ô 8 — Xác định cột mục tiêu

target_column = "Price"

print("Target column:", target_column)
```

```python
# Ô 9 — Phân loại đặc trưng số và phân loại

numerical_features = df.select_dtypes(include=["int64", "float64"]).columns
categorical_features = df.select_dtypes(include=["object"]).columns

print("Numerical Features:")
print(numerical_features)

print("\nCategorical Features:")
print(categorical_features)
```

```python
# Ô 10 — Kiểm tra dữ liệu trùng lặp

duplicate_rows = df.duplicated().sum()

print("Duplicate rows:", duplicate_rows)
```

```python
# Ô 11 — Kiểm tra outlier của giá nhà

Q1 = df["Price"].quantile(0.25)
Q3 = df["Price"].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[
    (df["Price"] < lower_bound) |
    (df["Price"] > upper_bound)
]

print("Number of outliers:", outliers.shape[0])
```

```python
# Ô 12 — Histogram phân phối giá nhà

plt.figure(figsize=(10,6))

sns.histplot(df["Price"], bins=50, kde=True)

plt.title("Distribution of House Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")

plt.show()
```

```python
# Ô 13 — Boxplot giá nhà

plt.figure(figsize=(10,5))

sns.boxplot(x=df["Price"])

plt.title("Boxplot of House Prices")

plt.show()
```

```python
# Ô 14 — Chọn các cột số

numeric_df = df.select_dtypes(include=["int64", "float64"])

numeric_df.head()
```

```python
# Ô 15 — Heatmap tương quan

plt.figure(figsize=(12,8))

correlation_matrix = numeric_df.corr()

sns.heatmap(
    correlation_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Heatmap")

plt.show()
```

```python
# Ô 16 — Scatter plot diện tích và giá nhà

plt.figure(figsize=(10,6))

sns.scatterplot(
    x=df["Area"],
    y=df["Price"]
)

plt.title("Area vs Price")
plt.xlabel("Area")
plt.ylabel("Price")

plt.show()
```

```python
# Ô 17 — Scatter plot số phòng ngủ và giá nhà

plt.figure(figsize=(10,6))

sns.scatterplot(
    x=df["Bedrooms"],
    y=df["Price"]
)

plt.title("Bedrooms vs Price")
plt.xlabel("Bedrooms")
plt.ylabel("Price")

plt.show()
```

```python
# Ô 18 — Histogram diện tích nhà

plt.figure(figsize=(10,6))

sns.histplot(df["Area"], bins=40, kde=True)

plt.title("Distribution of Area")
plt.xlabel("Area")
plt.ylabel("Frequency")

plt.show()
```

```python
# Ô 19 — Boxplot diện tích nhà

plt.figure(figsize=(10,5))

sns.boxplot(x=df["Area"])

plt.title("Boxplot of Area")

plt.show()
```
